```python
weights=[[0.2,0.8,-0.5,1],[0.5,-0.91,0.26,-0.5],[-0.1,0.4,0.5,0.9]]
biases=[2,1,3]

#zip combines two lists into list of lists element wise
layer_outputs=[]#op of current layer
for neuron_weights,neuron_bias in zip(weights,biases):
  neuron_output=0 #output of given neuron
  for n_input,weight in zip(inputs,neuron_weights):
    neuron_output+=n_input*weight
  neuron_output+=neuron_bias # Add bias once per neuron
  layer_outputs.append(neuron_output)
print(layer_outputs)
```

```python
#using dot prod
output=np.dot(weights,inputs)+biases #the first el we pass is how the return is to be indexed,like how do we know e had three neurons,yes,we had a LOL of 3 sets of weights
print(output)
```

```python
X=[[0.2,0.8,-0.5,1],[0.5,-0.91,0.26,-0.5],[-0.1,0.4,0.5,0.9]]
np.random.seed(0)#start randomness from 0

class Layer_Dense: #without self its a memory game
  def __init__(self,n_inputs,n_neurons): #is the constructor method in Python, used to initialize a class instance's attributes automatically when an object is created. The __init__ (dunder init) method acts as a constructor, with self representing the specific object being created, allowing it to store data
    self.weights=0.10*np.random.randn(n_inputs,n_neurons)
    #earlier it was weight by inputs,we did this inputs by weights to not do transpose again in forward pass
    #what is the size of inputs coming n,and how many neurons,#randn in gauss distb bounded around 0
    self.biases=np.zeros((1,n_neurons))#shape is 1  by however neurons we have,since we have a single bias for each neuron eg 3 neuron=[0,0,0],also the tuple of shape is a param
     #whatever code inside init automatically runs when obj is created
  def forward(self,inputs):
    self.output=np.dot(inputs,self.weights)+self.biases
layer1=Layer_Dense(4,5)
layer2=Layer_Dense(5,2)
layer1.forward(X) #layer1 op
print(layer1.output)
layer2.forward(layer1.output)#layer 1 op becomes layer 2 ip
print(layer2.output)
```
